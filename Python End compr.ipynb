1. Distribution of employees

```python
teams = df["Team"].value_counts()
team_perc = (teams/len(df)) * 100

print(teams)
print(" ")
print(team_perc)
```

```python
# Graph

import matplotlib.pyplot as plt

teams.plot(kind = "bar",color = "red")
plt.xlabel("Teams")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution By Team")
plt.show()
```

some teams have more employees than others,showing uneven workforce distribution.

2. Segregate employees based on their positions within the company. (2 marks)

```python
position = df["Position"].value_counts()
print(position)
```

```python
import matplotlib.pyplot as plt

position.plot(kind = "bar",color = "green")
plt.xlabel("Position")
plt.ylabel("Count")
plt.title("Employees by Position")
plt.show()
```

Certain  job positions have a higher number of employees compared to others,

these roles are more common within the organization.

3. Identify the predominant age group among employees. (2 marks)

```python
df["Age Group"] = pd.cut(df["Age"],
bins = [20,25,30,35,40,45,50],
labels=["20-25","25-30","30-35","35-40","40-45","45-50"])

group_count = df["Age Group"].value_counts()

print(group_count)
```

```python
import matplotlib.pyplot as plt

group_count.plot(kind ="bar")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

The majority of employees belong to the  mid - age group(around 20-35 years).

This indicates that the company has  a relatively young and active workforce.

4. Discover which team and position have the highest salary expenditure. (2 marks)

```python
salary = df.groupby("Team")["Salary"].sum()
positions = df.groupby("Position")["Salary"].sum()

print(salary.idxmax(),salary.max())
print(positions.idxmax(),positions.max())
```

```python
salary.plot(kind = "bar")
plt.title("Salary Expenditure By Team")
plt.show()
```

```python
positions.plot(kind = "bar")
plt.title("Salary Expenditure By Position")
plt.show()
```

Certain teams and job positions have a higher total salary expenditure compared to others. 

The company is spending more money on these areas.

5. Investigate if there's any correlation between age and salary, and represent it visually. (2 marks)

```python
correlation = df["Age"].corr(df["Salary"])
print(correlation)
```

```python
plt.scatter(df["Age"],df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```

If the correlation is positive, it means that as age increases, salary also tends to increase.

If the correlation is weak or close to zero, it means there is little or no relationship between age and salary.
